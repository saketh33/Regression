# Train - Test - split

```python
#spliting them on base of null values in clear_date
train=df[df.clear_date.notnull()]
test=df[df.clear_date.isnull()]
```

```python
# changing it to datetime format
train[["clear_date"]] = train[["clear_date"]].apply(pd.to_datetime)
```

```python
train['delay']=train['clear_date']-train['due_in_date']
train['delay']=train['delay'].astype('str')
train['delay']=train['delay'].apply(lambda x: x.split()[0])

#changing the string type to int
train['delay']=train['delay'].astype('int')
```

# Feature Engineering

### business_code

```python
#encoding business_code using dummies
trainbusdum= pd.get_dummies(train.business_code)
testbusdum=pd.get_dummies(test.business_code)

train=pd.concat([train,trainbusdum], axis=1)
train.drop(columns='business_code',inplace=True,axis=1)

test=pd.concat([test,testbusdum], axis=1)
test.drop(columns='business_code',inplace=True,axis=1)
```

### dropping the constant columns

```python
constants= []
uniquevals= []
unique_cols = dict()
for i in list(train.columns):
    uniquevals.append(train[i].nunique())
    if(train[i].nunique()==1):
        constants.append(i)
print('Constant Features are : ',constants)
# Removing the constant feature from train
train=train.drop(columns=constants)
# Removing the constant feature from test
test=test.drop(columns=constants)
```

```python
train.info()
```

```python
# changing data from unit to int
def change(col):
    train[col]=train[col].astype('int64')
l=['CA02','U001','U002','U005','U007','U013']
for i in l:
    change(i)
```

```python
def change(col):
    test[col]=test[col].astype('int64')
l=['CA02','U001','U002','U005','U013']
for i in l:
    change(i)
```

```python
# adding this constant null column beacuse its in train
test['U007']=0
```

#### dropping unwanted columns from the observations of EDA

### cust_number,invoice_currency,cust_payment_terms

```python
# using label encoder for encoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
labels= ['cust_number','invoice_currency','cust_payment_terms']
def labelencoding(col):
    train[col]= labelencoder.fit_transform(train[col])
    test[col]= labelencoder.fit_transform(test[col])
for _ in labels:
    labelencoding(_)
```

```python
#encoding them to year,month,day split
train['create_year']=train['document_create_date.1'].dt.year
train['create_month']=train['document_create_date.1'].dt.month
train['create_days']=train['document_create_date.1'].dt.day

train['due_year']=train['due_in_date'].dt.year
train['due_month']=train['due_in_date'].dt.month
train['due_days']=train['due_in_date'].dt.day

test['create_year']=test['document_create_date.1'].dt.year
test['create_month']=test['document_create_date.1'].dt.month
test['create_days']=test['document_create_date.1'].dt.day

test['due_year']=test['due_in_date'].dt.year
test['due_month']=test['due_in_date'].dt.month
test['due_days']=test['due_in_date'].dt.day
```

# Train - Valid - Split

```python
#dividing into target and remaining columns
X=train.drop(columns=['delay'],axis=1)
y=train.delay
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xvalid, ytrain, yvalid= train_test_split(X, y, test_size=0.20, random_state=36)
```

# Modelling and Predicting

```python
import xgboost as xgb
xgb= xgb.XGBRegressor()
xgb.fit(Xtrain, ytrain)

# Predicting the Validation Set Results
xgbpred= xgb.predict(Xvalid)
```

```python
from sklearn.metrics import mean_squared_error
print('rmse',mean_squared_error(yvalid,xgbpred,squared=False))
from sklearn.metrics import r2_score
print('r2score',r2_score(yvalid,xgbpred))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_model = RandomForestRegressor(n_estimators=120)
random_model.fit(Xtrain, ytrain)
y_pred3 = random_model.predict(Xvalid)
from sklearn.metrics import mean_squared_error
print("rmse",mean_squared_error(yvalid,y_pred3,squared=False))
from sklearn.metrics import r2_score
print("r2",r2_score(yvalid,y_pred3))
random_model_accuracy = round(random_model.score(Xtrain,ytrain)*100,2)
print("score",round(random_model_accuracy,2),'%')
```
